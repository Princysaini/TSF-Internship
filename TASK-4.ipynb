# TASK 4 - Decision Tree Algorithm

#### A decision tree is a kind of machine learning algorithm that can be used for classification or regression.

```python
# Importing Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn.datasets as ds
%matplotlib inline
from sklearn.model_selection import train_test_split

data = ds.load_iris()
data1 = pd.DataFrame(data.data,columns = data.feature_names)
print(data1.head())
y=data.target
print(y)
```

```python
from sklearn.tree import DecisionTreeClassifier
md = DecisionTreeClassifier(min_samples_split=5)
md.fit(data1,y)
print('Decision Tree Classifer Created')
```

```python
from sklearn import tree
plt.figure(figsize = (20,20))
tree.plot_tree(md , filled = True)
plt.title("Decisionn Tree of Iris Data")
```
